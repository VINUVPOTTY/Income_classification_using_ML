```python
from scipy.stats import chi2_contingency
for col in ['workclass', 'education', 'marital-status', 'occupation', 'relationship','race','sex','native-country']:
    table = pd.crosstab(df[col], df['salary'])
    _, p, _, _ = chi2_contingency(table)
    print(f"{col}: p-value = {p:.4f} â†’ {'Significant' if p < 0.05 else 'Not Significant'}")
```

```python
for column in df:
    plt.figure(figsize=(12,6))
    sns.boxplot(y=column,data=df)
    plt.show()
```

```python

for col in df.select_dtypes(include='number').columns:
    if col == 'salary':
        continue 
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
dfe=pd.get_dummies(df[['workclass', 'education', 'occupation']],drop_first=True,dtype=int)
dfe
```

```python
df_1=pd.concat([dfe,df],axis=True)
```

```python
df_1.drop(['workclass', 'education','occupation'],axis=1,inplace=True)
```

```python
df_1
```

```python
x=df_1.drop(['salary'],axis=1)
x
```

```python
y=df_1['salary']
y
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x,y)
```

```python

sns.countplot(x=y_resampled, palette="coolwarm")
plt.title("Balanced Target Distribution After SMOTE")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_resampled,y_resampled,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
knn=KNeighborsClassifier(n_neighbors=7)
naive=BernoulliNB()
support=SVC()
lst=[knn,naive,support]
```

```python
for i in lst:
    i.fit(x_train,y_train)
    y_predict=i.predict(x_test)
    print("performance evaluation of :- ",i)
    print("confusion_matrix is :-",confusion_matrix(y_test,y_predict))
    print("*"*100)
    print("accuracy score is :-",accuracy_score(y_test,y_predict))
    print("*"*100)
    print("classification report is :-",classification_report(y_test,y_predict))
    print("*"*100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

for i in lst:
    i.fit(x_train, y_train)
    y_predict = i.predict(x_test)
    ConfusionMatrixDisplay.from_predictions(y_test, y_predict)
    plt.title(f"Confusion Matrix - {i.__class__.__name__}")
    plt.show()
```
